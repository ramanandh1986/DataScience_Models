```python
## Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
## Reading Dataset
filepath = r"C:\Users\rayad\Documents\Data Science_Assignments\Recommendation_System\book.csv"
df = pd.read_csv(filepath, encoding='latin-1')
df.head()
```

```python
## Removing Unnamed Column
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
## Renaming Columns
df.columns = ['UserID', 'Title', 'Rating']
df.head()
```

```python
df.info()
```

```python
## Creating a Pivot Table
bookmat = df.pivot_table(index = 'UserID', columns= 'Title', values = 'Rating')
UserID = bookmat.index
Title = bookmat.columns
print(bookmat.shape)
bookmat.head()
```

```python
bookmat.fillna(0, inplace=True)
bookmat.head()
```

```python
## Importing Libraries to Calculate the similarities
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
# Finding the similarities
book_sim = 1-pairwise_distances(bookmat.values, metric='cosine')
book_sim
```

```python
## Storing the results in the dataframe
book_sim_df = pd.DataFrame(book_sim)
book_sim_df.head()
```

```python
## Setting Index and Columnn
book_sim_df.index = df.UserID.unique()
book_sim_df.columns = df.UserID.unique()
```

```python
book_sim_df
```

```python
# Filling Diagonal with '0'
np.fill_diagonal(book_sim, 0)
book_sim_df.head(20)
```

```python
book_sim_df.idxmax(axis=1)[0:20]
```

```python
count_of_ratings = pd.DataFrame(df.groupby('Title')['Rating'].count().sort_values(ascending = False))
count_of_ratings.head(10)
```

```python
average_ratings = pd.DataFrame(df.groupby('Title')['Rating'].mean().sort_values(ascending = False))
average_ratings.head(10)
```

```python
sum_of_ratings = pd.DataFrame(df.groupby('Title')['Rating'].sum().sort_values(ascending = False))
sum_of_ratings.head(10)
```

```python
book_rating_df = sum_of_ratings.merge(count_of_ratings, on='Title').merge(average_ratings, on='Title')
book_rating_df.head()
```

```python
book_rating_df.columns = ['Total Rating', 'Rated_By', 'Average Rating']
book_rating_df.sort_values('Total Rating', ascending= False).head(10)
```

```python
plt.figure(figsize=(6,4))
book_rating_df['Rated_By'].hist()
```

```python
plt.figure(figsize=(6,4))
book_rating_df['Average Rating'].hist()
```

```python
sns.jointplot(x= 'Average Rating', y='Rated_By', data= book_rating_df, alpha = 0.5)
```

```python
sns.jointplot(x= 'Total Rating', y='Rated_By', data= book_rating_df, alpha = 0.5)
```
