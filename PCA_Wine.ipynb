```python
## Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
## Reading Dataset

wine = pd.read_csv('wine.csv')

wine.head()
```

```python
wine.isna().sum()
```

```python
wine.shape
```

```python
wine_new = wine.iloc[:, 1:]
wine_new.head()
```

```python
## Converting into an array
wine_df = wine_new.values
wine_df
```

```python
## Normalizing the data
wine_norm = scale(wine_df)
wine_norm
```

```python
## Initializing PCA
pca = PCA(n_components= 3)
pca_values = pca.fit_transform(wine_norm)
pca_values
```

```python
## Creating Dataframe for top 3 PCA values
pca_df = pd.DataFrame(pca_values, columns=['PCA1', 'PCA2', 'PCA3'])
pca_df
```

```python
pca.components_
```

```python
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], c = wine['Type'])
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```

```python
var = pca.explained_variance_ratio_
var
```

```python
## Cumulative Sum of top 3 variances
var_cumulative = np.cumsum(np.round(var, decimals= 3)*100)
var_cumulative
```

```python
wine_final = pd.concat([pd.DataFrame(pca_values[:, 0:3], columns= ['PCA1', 'PCA2', 'PCA3']), wine[['Type']]], axis=1)
wine_final
```

## Hierarchical Clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
import seaborn as sns
from scipy.cluster.hierarchy import linkage
```

```python
df = pca_df
df.head()
```

```python
## Creating Dendrogram
x = linkage(df, method='ward', metric= 'euclidean')
plt.figure(figsize=(20,8))
plt.title('Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(x,
              leaf_rotation=0,
              leaf_font_size=5,
              show_contracted= True)
plt.show()
```

```python
h_pca = AgglomerativeClustering(n_clusters=3, linkage='complete', affinity='euclidean').fit(pca_df)
cluster_labels = pd.Series(h_pca.labels_)
pca_df['Cluster'] = cluster_labels
```

## K-Mean Clustering

```python
df_k = pca_df.iloc[:, 0:3]
df_k
```

```python
## Within Cluster Sum of Squares
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters = i, random_state = 1)
    kmeans.fit(df_k)
    wcss.append(kmeans.inertia_)
    
wcss
```

```python
## Plotting Elbow Curve
plt.plot(range(1,11), wcss)
plt.title('Elbow Curve')
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
## Identifying the Optimal K Value
from kneed import KneeLocator

y = wcss
x = range(1, len(y)+1)

kn = KneeLocator(x, y, curve= 'convex', direction='decreasing')
print(kn.knee)
```

```python
plt.plot(x, y, 'bx--')
plt.xlabel('Number of Clusters')
plt.ylabel('Distances')
plt.vlines(kn.knee, plt.ylim()[0], plt.xlim()[1], linestyles='dotted')
plt.show()
```

```python
Clusters_k = KMeans(3, random_state= 12)
Clusters_k.fit(df_k)
```

```python
Clusters_k.labels_
```

```python
df_k['K Clusters'] = Clusters_k.labels_
df_k
```
